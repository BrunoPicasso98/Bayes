```python
import numpy as np
import pandas as pd
from scipy.stats import uniform
import seaborn as sns
```

```python
def toss_a_coin(): 
    if np.random.rand() > 0.5:
        return 1
    else:
        return 0
```

```python
toss_a_coin()
```

```python
n = 1000
start = 10 
width = 20
data = uniform.rvs(size = n, loc = start, scale = width)
```
